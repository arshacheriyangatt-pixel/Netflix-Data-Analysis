```python
import pandas as pd

def categorize_rating(rating):
    if pd.isna(rating):
        return "Unknown"
    
    rating = str(rating).strip()

    # Handle wrong entries like durations
    if "min" in rating:
        return "Unknown"
    
    # Kids content
    if rating in ["TV-Y", "TV-Y7", "TV-Y7-FV", "G"]:
        return "Kids"
    
    # Family / General
    elif rating in ["TV-G", "PG", "TV-PG"]:
        return "Family"
    
    # Teens
    elif rating in ["PG-13", "TV-14"]:
        return "Teen"
    
    # Adults
    elif rating in ["R", "TV-MA", "NC-17"]:
        return "Adult"
    
    # Not Rated / Others
    elif rating in ["NR", "UR"]:
        return "Unrated"
    
    else:
        return "Unrated"
```

```python
df["rating_category"] = df["rating"].apply(categorize_rating)
```
